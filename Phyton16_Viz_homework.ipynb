```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
df = pd.read_csv('Happy/2019.csv')
df
```

```python
df.info()
```

```python
#Тренировка
plt.plot(df['GDP per capita'], df.Score)
plt.title('Зависимость счастья от уровня ВВП на душу населения')
plt.xlabel('ВВП на душу')
plt.ylabel('Индекс счастья')
#График показывает, что чем выше урвень жизни, тем, как правило, люди более счастливые
# А еще показывает как не надо строить график
# не знал, что каракули ребенка - это зависимость индекса счастья от ВВП
# А еще интересный факт, что есть страны,имеющие одинаковый индекс счастья при значительной разнице ВВП 
#- на коротком учстке не в деньгах счастье, но в целом в них, если они есть
```

```python
#то же самое вторым способом ,более наглядно
df.plot(title='Зависимость счастья от уровня ВВП на душу населения', 
        x='GDP per capita', 
        y='Score',
        kind='scatter')
```

```python
# то же самое в seaborn
#Наглядно представлена зависимость счастья от уровня ВВП
ax = sns.regplot(data=df, x='GDP per capita', y='Score', color='green', marker='+')
ax.set_title('Зависимость счастья от уровня ВВП на душу населения')
```

```python
# то же самое в seaborn
#Наглядно представлена зависимость счастья от уровня социалки
ax = sns.regplot(data=df, x='Social support', y='Score', color='green', marker='+')
ax.set_title('Зависимость счастья от социалки')
```

```python
# то же самое в seaborn
#Наглядно представлена зависимость счастья от уровня здоровья
ax = sns.regplot(data=df, x='Healthy life expectancy', y='Score', color='green', marker='+')
ax.set_title('Зависимость счастья от здоровья')
```

```python
# то же самое в seaborn
#Видим большой разброс, т.е свобода выбора в меньшей степени влияет на ощущение счастья по сравнению с экономикой, социалкой и здравоохранением
ax = sns.regplot(data=df, x='Freedom to make life choices', y='Score', color='green', marker='+')
ax.set_title('Зависимость счастья от свободы выбора')
```

```python
#Теперь посмотрим Счасливы ли люди в мире, как много счастливых стран
df.Score.plot(kind='hist', 
                     bins=11,
                     title='Распределение счастья в мире',
                     
                     xticks=range(2, 9))
#Больниство стран живут со средним уровнем счастья, очень счастливых и грустных стран примерно поровну
```
