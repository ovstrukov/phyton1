```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("vgsales.csv") 
df = df.dropna()
df
```

```python
#1) Как критики относятся к спортивным играм?

df_sport = df[df.Genre == 'Sports']
df_sport
```

H0 - средняя оценка критиков спортивных игр >= 70%
H1 - средняя оценка критиков спортивных игр < 70%

```python
alpha = 0.05
result = st.ttest_1samp(df_sport['Critic_Score'], 70, alternative='less')
print(df_sport['Critic_Score'].mean())
print(result)
if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, среднее меньше 50, критики негативно оценивают спортивные игры')
else:
    print('Не отвергаем нулевую гипотезу, среднее больше 50, критики положительно оценивают спортивные игры')
```

```python
#2) Критикам нравятся больше игры на PC или на PS4?
df_PC = df[df.Platform == 'PC']
df_PS4 = df[df.Platform == 'PS4']
```

H0 - средняя оценка критиков PC и PS4 одинакова
H1 - средняя оценка критиков PC и PS4 отличается

```python
alpha = 0.05
result = st.ttest_ind(df_PC['Critic_Score'], df_PS4['Critic_Score'], equal_var = False)
print(df_PC['Critic_Score'].mean())
print(df_PS4['Critic_Score'].mean())
print(result)
if result.pvalue < alpha and result.statistic > 0:
    print('Отвергаем нулевую гипотезу, критики по разному оценивают PC и PS4, PC лучше оценивают на:', result.statistic)
elif result.pvalue < alpha and result.statistic < 0:
    print('Отвергаем нулевую гипотезу, критики по разному оценивают PC и PS4, PC хуже оценивают на:', result.statistic)
else:
    print('Не отвергаем нулевую гипотезу, критики одинаково оценивают PC и PS4')
```

3) Критикам больше нравятся стрелялки или стратегии?

```python
df_Shooter = df[df.Genre == 'Shooter']
df_Strategy = df[df.Genre == 'Strategy']
```

```python
H0 - средняя оценка критиков Shooter и Strategy одинакова
H1 - средняя оценка критиков Shooter и Strategy отличается
```

```python
alpha = 0.05
result = st.ttest_ind(df_Shooter['Critic_Score'], df_Strategy['Critic_Score'], equal_var = False)
print(df_Shooter['Critic_Score'].mean())
print(df_Strategy['Critic_Score'].mean())
print(result)
if result.pvalue < alpha and result.statistic > 0:
    print('Отвергаем нулевую гипотезу, критики по разному оценивают Shooter и Strategy, Shooter лучше оценивают на:', result.statistic)
elif result.pvalue < alpha and result.statistic < 0:
    print('Отвергаем нулевую гипотезу, критики по разному оценивают Shooter и Strategy, Shooter хуже оценивают на:', result.statistic)
else:
    print('Не отвергаем нулевую гипотезу, критики одинаково оценивают Shooter и Strategy')
```

```python
import pandas as pd
import re
```

```python
emails = pd.read_csv("spam.csv") 
texts = emails['Message']
texts[15]
emails
```

```python
# приведем все слова к нижнему регистру
texts_lower_case = [text.lower() for text in texts]
# проверим
texts_lower_case[1]
```

```python
# уберем все служебные символы
print(len(texts_lower_case[15]))
only_words_text = [re.sub('[\W_]+',' ', text) for text in texts_lower_case]
# проверим
print(len(only_words_text[55]))
only_words_text[1]
```

```python
# разделим все на слова
texts_splitted = [text.split() for text in only_words_text]
# сколько слов?
len(texts_splitted[1])
```

```python
#фильтруем стоп-слова
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
without_sw = [[word for word in text if word not in stopwords_set] for text in texts_splitted]
without_sw[1]
```

```python
#проводим лемматизацию
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_text = [[wordnet_lemmatizer.lemmatize(word) for word in text] for text in without_sw]
lemmatized_text[1]
```

```python
#склеиваем назад список слов с строки
LemMessage=[]
for item in lemmatized_text:
    LemMessage.append(" ".join(map(str, item)))
LemMessage[1]
```

```python
#перезаписываем сообщение лемматизированным и преобразованным текстом
emails['Message'] = LemMessage
emails
```

```python
#Преобразуйте все сообщения в вектора TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(emails.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
```

```python
#подключаем логистическую регрессию
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#задаем x, y
X = tfidf_matrix
y = emails['Category']
```

```python
# делим на тренировочную часть  и тестовую
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# формируем модель LR
model = LogisticRegression()
model.fit(X_train, y_train) 
```

```python
#Обучаем
model.fit(X_train, y_train) 
```

```python
#Предсказываем
y_pred = model.predict(X_test)
y_pred[1]
```

```python
#Опишите результаты при помощи confusion_matrix
#1445 писем модель верно предсказала спам
#66 писем модель посчитала как спам, но на самом деле это обычные письма (ошибка 1 рода)
#3 письма посчитала как не спам, но на самом деле это спам (ошибка 2 рода)
#158 писем модель определила как обычные и в действительности они не являются спамом
import sklearn as sklearn
sklearn.metrics.confusion_matrix(y_test, y_pred, labels=['spam', 'ham'], sample_weight=None, normalize=None)
```
